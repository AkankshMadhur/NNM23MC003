```python
import matplotlib.pyplot as plt

x = []
y = []

with open('test.txt', 'r', encoding='utf-8-sig') as file:
    for line in file:
        values = line.split()
        if len(values) == 2:  
            x.append(int(values[0]))
            y.append(int(values[1]))

# Plot the data
plt.plot(x, y, marker='o', color='b')
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Line Graph from Text File")
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from io import StringIO
data = """Date,Open,High,Low,Close
10-03-16,774.25,776.065002,769.5,772.559998
10-04-16,776.030029,778.710022,772.890015,776.429993
10-05-16,779.309998,782.070007,775.650024,776.469971
10-06-16,779,780.47998,775.539978,776.859985
10-07-16,779.659973,779.659973,770.75,775.080017
"""
df = pd.read_csv(StringIO(data), sep=',', parse_dates=['Date'], dayfirst=True)
df.set_index('Date', inplace=True)
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['Open'], marker='o', label='Open Price', color='blue')
plt.plot(df.index, df['High'], marker='o', label='High Price', color='green')
plt.plot(df.index, df['Low'], marker='o', label='Low Price', color='red')
plt.plot(df.index, df['Close'], marker='o', label='Close Price', color='orange')
plt.xlabel('Date')
plt.ylabel('Price (USD)')

plt.title('Financial Data of Alphabet Inc. (Oct 3, 2016 - Oct 7, 2016)')
plt.xticks(rotation=45)
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [0, 1, 2, 3, 4, 5]

line1_y = [0, 1, 4, 9, 16, 25]
line1_label = 'Line 1'
line1_color = 'blue'

line2_y = [0, 2, 1, 3, 2, 4]
line2_label = 'Line 2'
line2_color = 'red'

line3_y = [0, 1, 2, 1, 2, 3]
line3_label = 'Line 3'
line3_color = 'green'
fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(x, line1_y, marker='o', color=line1_color, label=line1_label)
ax.plot(x, line2_y, marker='o', color=line2_color, label=line2_label)
ax.plot(x, line3_y, marker='o', color=line3_color, label=line3_label)
ax.set_xlabel('X Axis Label')
ax.set_ylabel('Y Axis Label')
ax.set_title('Multiple Lines with Legends')
ax.legend()
ax.grid()
plt.tight_layout()
plt.show()
```

```python
#SIMPLE LINEAR REGRESSION

import numpy as np
import matplotlib.pyplot as plt

X = np.array([1, 2, 3, 4, 5, 6, 7])
Y = np.array([1.5, 3.8, 6.7, 9.0, 11.2, 13.6, 16])

n = len(X)

sumX = np.sum(X)
sumY = np.sum(Y)
sumXY = np.sum(X * Y)
sumX_square = np.sum(X ** 2)

m = (n * sumXY - sumX * sumY) / (n * sumX_square - sumX ** 2)
b = (sumY * sumX_square - sumX * sumXY) / (n * sumX_square - sumX ** 2)

Y_pred = m * X + b

print(f"Slope (m): {m}")
print(f"Intercept (b): {b}")
print(f"Predicted Y values: {Y_pred}")

plt.scatter(X, Y, color='blue', label='Original Data')
plt.plot(X, Y_pred, color='red', label='Predicted Line')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression')

plt.legend()
plt.show()
```
